```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Scaler
from sklearn.preprocessing import MinMaxScaler

# Importing the Keras libraries and packages
from keras.models import Model, Sequential
from keras.layers import Input, Dense

import pylab

import itertools

```

```python
stock_lst = ["Verizon", "AT&T", "T - Mobile", "Sprint"]

verizon_df = pd.read_csv("data/VZ.csv")
atat_df = pd.read_csv("data/T.csv")
sprint_df = pd.read_csv("data/S.csv")
t_mobile_df = pd.read_csv("data/TMUS.csv")

date_lst = verizon_df['Date'].values

snp500_df = pd.read_csv("data/SP500.csv")


interest_df = pd.read_csv("data/DFF.csv")
interest_df = interest_df[interest_df["DATE"].isin(date_lst)]

```

```python
# Extract adjusted price for each df

# Adj Close
verizon_adj_closed = verizon_df.iloc[:, 5:6].values
atat_adj_closed = atat_df.iloc[:, 5:6].values
sprint_adj_closed = sprint_df.iloc[:, 5:6].values
t_mobile_adj_closed = t_mobile_df.iloc[:, 5:6].values

snp500_adj_closed = snp500_df.iloc[:, 5:6].values

# Volume
verizon_volume = verizon_df.iloc[:, 6:7].values
atat_volume = atat_df.iloc[:, 6:7].values
sprint_volume = sprint_df.iloc[:, 6:7].values
t_mobile_volume = t_mobile_df.iloc[:, 6:7].values

# interest 
interest_rates = interest_df.iloc[:,1:2].values
```

```python
# Feature scaling

## Setting up scaler
# Adj Close
sc_verizon_adj_closed = MinMaxScaler(feature_range = (0, 1))
sc_atat_adj_closed = MinMaxScaler(feature_range = (0, 1))
sc_sprint_adj_closed = MinMaxScaler(feature_range = (0, 1))
sc_t_mobile_adj_closed = MinMaxScaler(feature_range = (0, 1))

sc_snp500_adj_closed = MinMaxScaler(feature_range = (0, 1))

# Volume
sc_verizon_volume = MinMaxScaler(feature_range = (0, 1))
sc_atat_volume = MinMaxScaler(feature_range = (0, 1))
sc_sprint_volume = MinMaxScaler(feature_range = (0, 1))
sc_t_mobile_volume = MinMaxScaler(feature_range = (0, 1))

# interest 
sc_interest_rates = MinMaxScaler(feature_range = (0, 1))


# scaling data

# Adj Close
verizon_adj_closed_scaled = sc_verizon_adj_closed.fit_transform(verizon_adj_closed)
atat_adj_closed_scaled = sc_atat_adj_closed.fit_transform(atat_adj_closed)
sprint_adj_closed_scaled = sc_sprint_adj_closed.fit_transform(sprint_adj_closed)
t_mobile_adj_closed_scaled = sc_t_mobile_adj_closed.fit_transform(t_mobile_adj_closed)

snp500_adj_closed_scaled = sc_snp500_adj_closed.fit_transform(snp500_adj_closed)

# Volume
verizon_volume_scaled = sc_verizon_volume.fit_transform(verizon_volume)
atat_volume_scaled = sc_atat_volume.fit_transform(atat_volume)
sprint_volume_scaled = sc_sprint_volume.fit_transform(sprint_volume)
t_mobile_volume_scaled = sc_t_mobile_volume.fit_transform(t_mobile_volume)

# interest 
interest_rates_scaled = sc_interest_rates.fit_transform(interest_rates)
```

```python
n_day = 30
n_predict = 5

X_verizon = []
y_verizon = []

X_atat = []
y_atat = []

X_sprint = []
y_sprint = []

X_t_mobile = []
y_t_mobile = []

for i in range(n_day, 502 - n_predict):
    # verizon
    X_tmp_verizon = np.concatenate((verizon_adj_closed_scaled[i-n_day:i, 0], verizon_volume_scaled[i-n_day:i, 0], 
                                    snp500_adj_closed_scaled[i-n_day:i, 0], interest_rates_scaled[i-n_day:i, 0]))
    y_tmp_verison = verizon_adj_closed_scaled[i:i+n_predict, 0]
    
    X_verizon.append(X_tmp_verizon)
    y_verizon.append(y_tmp_verison)

    # at-t
    X_tmp_atat = np.concatenate((atat_adj_closed_scaled[i-n_day:i, 0], atat_volume_scaled[i-n_day:i, 0], 
                                    snp500_adj_closed_scaled[i-n_day:i, 0], interest_rates_scaled[i-n_day:i, 0]))
    y_tmp_atat = atat_adj_closed_scaled[i:i+n_predict, 0]
    
    X_atat.append(X_tmp_atat)
    y_atat.append(y_tmp_atat)
    
    # sprint
    X_tmp_sprint = np.concatenate((sprint_adj_closed_scaled[i-n_day:i, 0], sprint_volume_scaled[i-n_day:i, 0], 
                                    snp500_adj_closed_scaled[i-n_day:i, 0], interest_rates_scaled[i-n_day:i, 0]))
    
    y_tmp_sprint = sprint_adj_closed_scaled[i:i+n_predict, 0]
    
    X_sprint.append(X_tmp_sprint)
    y_sprint.append(y_tmp_sprint)
    
    # t-mobile 
    X_tmp_t_mobile = np.concatenate((t_mobile_adj_closed_scaled[i-n_day:i, 0], t_mobile_volume_scaled[i-n_day:i, 0], 
                                    snp500_adj_closed_scaled[i-n_day:i, 0], interest_rates_scaled[i-n_day:i, 0]))
    y_tmp_t_mobile = t_mobile_adj_closed_scaled[i:i+n_predict, 0]
    
    X_t_mobile.append(X_tmp_t_mobile)
    y_t_mobile.append(y_tmp_t_mobile)
    

X_verizon, y_verizon = np.array(X_verizon), np.array(y_verizon)
X_atat, y_atat = np.array(X_atat), np.array(y_atat)
X_sprint, y_sprint = np.array(X_sprint), np.array(y_sprint)
X_t_mobile, y_t_mobile = np.array(X_t_mobile), np.array(y_t_mobile)
```

```python
def build_and_train_model(stock_title, X_data, y_data, scaler):
    # inputs
    inputs = Input(shape=(n_day*4,))

    # hidden layers
    h1 = Dense(30, activation='linear')(inputs)
    h2 = Dense(300, activation='relu')(h1)
    h3 = Dense(120, activation='linear')(h2)
    
    # Outputs
    outputs = Dense(n_predict, activation='relu')(h3)

    # compile model
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer = 'adam', loss='mse', metrics=['mse', 'mae', 'mape', 'cosine'])
    
    # Train model
    history = model.fit(X_data, y_data, epochs = 100, batch_size = 30, verbose=0)  # starts training

    predicted_stock_price = model.predict(X_data)
    predicted_stock_price = scaler.inverse_transform(predicted_stock_price)

    mse = history.history['mse']
    plt.plot(mse)
    plt.xlabel('epoch')
    plt.ylabel('mse')
    plt.title("%s\nlast mse = %.6f"%(stock_title, mse[-1]))

    plt.show()
    
    return model, predicted_stock_price
```

```python
verizon_model, verizon_predicted_stock_price = build_and_train_model("Verizon", X_verizon, y_verizon, 
                                                                     sc_verizon_adj_closed)
```

```python
atat_model, atat_predicted_stock_price = build_and_train_model("AT&T", X_atat, y_atat,
                                                              sc_atat_adj_closed)
```

```python
sprint_model, sprint_predicted_stock_price = build_and_train_model("Sprint", X_sprint, y_sprint,
                                                                  sc_sprint_adj_closed)
```

```python
t_mobile_model, t_mobile_predicted_stock_price = build_and_train_model("T-Mobile", X_t_mobile, y_t_mobile,
                                                                      sc_t_mobile_adj_closed)
```

```python
def color_gen(cmap = "Blues", num_col = 7):
    cm = pylab.get_cmap(cmap)

    return [cm(1.*i/num_col) for i in range(num_col-1,-1,-1)]
```

```python
def plot_training(stock_title, real_stock_price, predicted_stock_price):
    real_stock_price_padded = real_stock_price.copy()
    predicted_stock_price_padded = predicted_stock_price.copy()

    nan_pred = [np.NaN for i in range(n_predict)]

    for i in range(n_day):
        real_stock_price_padded = np.append(real_stock_price_padded, [np.NaN])
        predicted_stock_price_padded = np.insert(predicted_stock_price_padded, 0, nan_pred)

    predicted_stock_price_padded = predicted_stock_price_padded.reshape(-1, n_predict)

    plt.figure(figsize = (20, 5))

    plt.plot(real_stock_price_padded, color = 'red', label = 'Actual %s Stock Price' % stock_title)

    col_day = list((color_gen(cmap = "Blues", num_col = 10)))

    for i in range(n_predict-1, -1, -1):
    #     print(i)
        day_pad = [np.NaN for j in range(i)]
        predicted_stock_price_tmp =  day_pad + [p[i] for p in predicted_stock_price_padded]

        plt.plot(predicted_stock_price_tmp, '--', color = col_day[n_predict-1-i],# alpha = i*0.1, 
                 label = '{} day prediction'.format(n_predict - i))

    plt.xlabel("Day")
    plt.ylabel("Adjusted Close Price")
    #     plt.xticks(range(0, len(date_lst), 30), date_lst[0::30])

    plt.xticks(rotation=45)
    plt.legend()

    plt.show()
```

```python
plot_training('Verizon', verizon_adj_closed, verizon_predicted_stock_price)
```

```python
plot_training('AT&T', atat_adj_closed, atat_predicted_stock_price)
```

```python
plot_training('Sprint', sprint_adj_closed, sprint_predicted_stock_price)
```

```python
plot_training('T-Mobile', t_mobile_adj_closed, t_mobile_predicted_stock_price)
```

```python
def get_percentage(p1, p2):
    return 100*(p2 - p1)/p1

def plot_test(stock_title, adj_closed, adj_closed_scaled, volume_scaled, scaler):
    
    # Prediction
    X_test = np.concatenate((adj_closed_scaled[-n_day:, 0], volume_scaled[-n_day:, 0], 
                                    snp500_adj_closed_scaled[-n_day:, 0], interest_rates_scaled[-n_day:, 0]))

    X_test = X_test.reshape(1,-1)

    real_stock_price = adj_closed[-n_day:, 0]

    predicted_stock_price = model.predict(X_test)
    predicted_stock_price = scaler.inverse_transform(predicted_stock_price)

    plt.figure(figsize = (20, 5))

    plt.plot(real_stock_price, color = 'red', label = 'Real %s Stock Price'% stock_title)

    col_day = list((color_gen(cmap = "Blues", num_col = 10)))

    for day, pred_price in enumerate(predicted_stock_price[0]):
        plt.scatter(n_day + day, pred_price, color = col_day[day],
                   label = '{} day prediction'.format(day+1))

    read_pred_tmp =  np.insert(predicted_stock_price[0], 0, real_stock_price[-1])

    plt.plot([n_day - 1 + day for day in range(n_predict + 1)], read_pred_tmp, '--', color = 'k')

    plt.xlabel("Day")
    plt.ylabel("Adjusted Close Price")
    
    pred_date_lst = ['2019-12-%d'%(13+i) for i in range(n_predict)]
    date_lst_test = itertools.chain(date_lst[-30:], pred_date_lst)
    date_lst_test = list(date_lst_test)

    plt.xticks(range(35), date_lst_test)

    plt.xticks(rotation=90)
    plt.legend()
    plt.title("%s Test Predictions" % stock_title)
    plt.show()
    
    # Price percentage
    p_lst = []
    bar_col = []
    num_bars = len(read_pred_tmp) - 1
    
    for i in range(num_bars):
        p = get_percentage(read_pred_tmp[i], read_pred_tmp[i+1])
        p_lst.append(p)

        col = 'r' if p < 0 else 'b'
        bar_col.append(col)

    plt.ylabel("Percent Increased")
    plt.xlabel("Date")
    plt.bar(range(num_bars), p_lst, color = bar_col)

    plt.xticks(range(n_predict), pred_date_lst)
    
    plt.title("%s Test Prediction's Percentage" % stock_title)
    plt.show()
```

```python
plot_test("Verizon", verizon_adj_closed, verizon_adj_closed_scaled, verizon_volume_scaled, sc_verizon_adj_closed)
```

```python
plot_test("AT&T", atat_adj_closed, atat_adj_closed_scaled, atat_volume_scaled, sc_atat_adj_closed)
```

```python
plot_test("Sprint", sprint_adj_closed, sprint_adj_closed_scaled, sprint_volume_scaled, sc_sprint_adj_closed)
```

```python
plot_test("T-Mobile", t_mobile_adj_closed, t_mobile_adj_closed_scaled, t_mobile_volume_scaled, sc_t_mobile_adj_closed)
```
