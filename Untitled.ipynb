```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Scaler
from sklearn.preprocessing import MinMaxScaler

# Importing the Keras libraries and packages
from keras.models import Sequential
from keras.layers import Dense, Activation
# from keras.layers import LSTM
from keras.layers import Dropout

import pylab
```

```python
def color_gen(cmap = "Blues", num_col = 7):
    cm = pylab.get_cmap(cmap)

    return [cm(1.*i/num_col) for i in range(num_col)]
```

```python
stock_lst = ["Verizon", "AT&T", "T - Mobile", "Sprint"]

verizon_df = pd.read_csv("data/VZ.csv")
atat_df = pd.read_csv("data/T.csv")
sprint_df = pd.read_csv("data/S.csv")
t_mobile_df = pd.read_csv("data/TMUS.csv")

date_lst = verizon_df['Date'].values

snp500_df = pd.read_csv("data/SP500.csv")


interest_df = pd.read_csv("data/DFF.csv")
interest_df = interest_df[interest_df["DATE"].isin(date_lst)]

```

```python
# Extract adjusted price for each df

# Adj Close
verizon_adj_closed = verizon_df.iloc[:, 5:6].values
atat_adj_closed = atat_df.iloc[:, 5:6].values
sprint_adj_closed = sprint_df.iloc[:, 5:6].values
t_mobile_adj_closed = t_mobile_df.iloc[:, 5:6].values

snp500_adj_closed = snp500_df.iloc[:, 5:6].values

# Volume
verizon_volume = verizon_df.iloc[:, 6:7].values
atat_volume = atat_df.iloc[:, 6:7].values
sprint_volume = sprint_df.iloc[:, 6:7].values
t_mobile_volume = t_mobile_df.iloc[:, 6:7].values

# interest 
interest_rates = interest_df.iloc[:,1:2].values
```

```python
# Feature scaling

## Setting up scaler
# Adj Close
sc_verizon_adj_closed = MinMaxScaler(feature_range = (0, 1))
sc_atat_adj_closed = MinMaxScaler(feature_range = (0, 1))
sc_sprint_adj_closed = MinMaxScaler(feature_range = (0, 1))
sc_t_mobile_adj_closed = MinMaxScaler(feature_range = (0, 1))

sc_snp500_adj_closed = MinMaxScaler(feature_range = (0, 1))

# Volume
sc_verizon_volume = MinMaxScaler(feature_range = (0, 1))
sc_atat_volume = MinMaxScaler(feature_range = (0, 1))
sc_sprint_volume = MinMaxScaler(feature_range = (0, 1))
sc_t_mobile_volume = MinMaxScaler(feature_range = (0, 1))

# interest 
sc_interest_rates = MinMaxScaler(feature_range = (0, 1))


# scaling data

# Adj Close
verizon_adj_closed_scaled = sc_verizon_adj_closed.fit_transform(verizon_adj_closed)
atat_adj_closed_scaled = sc_atat_adj_closed.fit_transform(atat_adj_closed)
sprint_adj_closed_scaled = sc_sprint_adj_closed.fit_transform(sprint_adj_closed)
t_mobile_adj_closed_scaled = sc_t_mobile_adj_closed.fit_transform(t_mobile_adj_closed)

snp500_adj_closed_scaled = sc_snp500_adj_closed.fit_transform(snp500_adj_closed)

# Volume
verizon_volume_scaled = sc_verizon_volume.fit_transform(verizon_volume)
atat_volume_scaled = sc_atat_volume.fit_transform(atat_volume)
sprint_volume_scaled = sc_sprint_volume.fit_transform(sprint_volume)
t_mobile_volume_scaled = sc_t_mobile_volume.fit_transform(t_mobile_volume)

# interest 
interest_rates_scaled = sc_interest_rates.fit_transform(interest_rates)
```

```python
n_day = 30
n_predict = 7

X_verizon = []
y_verizon = []

X_atat = []
y_atat = []

X_sprint = []
y_sprint = []

X_t_mobile = []
y_t_mobile = []

for i in range(n_day, 502 - n_predict):
    # verizon
    X_tmp_verizon = np.concatenate((verizon_adj_closed_scaled[i-n_day:i, 0], verizon_volume_scaled[i-n_day:i, 0], 
                                    snp500_adj_closed_scaled[i-n_day:i, 0], interest_rates_scaled[i-n_day:i, 0]))
    y_tmp_verison = verizon_adj_closed_scaled[i:i+n_predict, 0]
    
    X_verizon.append(X_tmp_verizon)
    y_verizon.append(y_tmp_verison)

    # at-t
    X_tmp_atat = np.concatenate((atat_adj_closed_scaled[i-n_day:i, 0], atat_volume_scaled[i-n_day:i, 0], 
                                    snp500_adj_closed_scaled[i-n_day:i, 0], interest_rates_scaled[i-n_day:i, 0]))
    y_tmp_atat = atat_adj_closed_scaled[i:i+n_predict, 0]
    
    X_atat.append(X_tmp_atat)
    y_atat.append(y_tmp_atat)
    
    # sprint
    X_tmp_sprint = np.concatenate((sprint_adj_closed_scaled[i-n_day:i, 0], sprint_volume_scaled[i-n_day:i, 0], 
                                    snp500_adj_closed_scaled[i-n_day:i, 0], interest_rates_scaled[i-n_day:i, 0]))
    
    y_tmp_sprint = sprint_adj_closed_scaled[i:i+n_predict, 0]
    
    X_sprint.append(X_tmp_sprint)
    y_sprint.append(y_tmp_sprint)
    
    # t-mobile 
    X_tmp_t_mobile = np.concatenate((t_mobile_adj_closed_scaled[i-n_day:i, 0], t_mobile_volume_scaled[i-n_day:i, 0], 
                                    snp500_adj_closed_scaled[i-n_day:i, 0], interest_rates_scaled[i-n_day:i, 0]))
    y_tmp_t_mobile = t_mobile_adj_closed_scaled[i:i+n_predict, 0]
    
    X_t_mobile.append(X_tmp_t_mobile)
    y_t_mobile.append(y_tmp_t_mobile)
    

X_verizon, y_verizon = np.array(X_verizon), np.array(y_verizon)
X_atat, y_atat = np.array(X_atat), np.array(y_atat)
X_sprint, y_sprint = np.array(X_sprint), np.array(y_sprint)
X_t_mobile, y_t_mobile = np.array(X_t_mobile), np.array(y_t_mobile)

# #reshaping input
# X_verizon = np.reshape(X_verizon, (X_verizon.shape[0], X_verizon.shape[1], 1))
# X_atat = np.reshape(X_atat, (X_atat.shape[0], X_atat.shape[1], 1))
# X_sprint = np.reshape(X_sprint, (X_sprint.shape[0], X_sprint.shape[1], 1))
# X_t_mobile = np.reshape(X_t_mobile, (X_t_mobile.shape[0], X_t_mobile.shape[1], 1))

# # reshaping output
# y_verizon = np.reshape(y_verizon, (y_verizon.shape[0], y_verizon.shape[1], 1))
# y_atat = np.reshape(y_atat, (y_atat.shape[0], y_atat.shape[1], 1))
# y_sprint = np.reshape(y_sprint, (y_sprint.shape[0], y_sprint.shape[1], 1))
# y_t_mobile = np.reshape(y_t_mobile, (y_t_mobile.shape[0], y_t_mobile.shape[1], 1))
```

```python
from keras.layers import Input, Dense
from keras.models import Model

# This returns a tensor
inputs = Input(shape=(120,))

# a layer instance is callable on a tensor, and returns a tensor
output_1 = Dense(64, activation='relu')(inputs)
output_2 = Dense(64, activation='relu')(output_1)
predictions = Dense(7, activation='softmax')(output_2)

# This creates a model that includes
# the Input layer and three Dense layers
model = Model(inputs=inputs, outputs=predictions)
model.compile(optimizer = 'adam', loss='mse', metrics=['mse', 'mae', 'mape', 'cosine'])

model.fit(X_verizon, y_verizon, epochs = 100, batch_size = 32)  # starts training

real_stock_price = verizon_adj_closed

predicted_stock_price = model.predict(X_verizon)
predicted_stock_price = sc_verizon_adj_closed.inverse_transform(predicted_stock_price)
```

```python
plt.figure(figsize = (20, 5))

plt.plot(real_stock_price, color = 'red', label = 'Real TATA Stock Price')

col_day = color_gen()
for i in range(n_predict):
    predicted_stock_price_tmp = [p[0] for p in predicted_stock_price]

    plt.plot(predicted_stock_price_tmp, color = col_day[i], label = '{} day prediction'.format(i+1))

plt.xlabel("Day")
plt.ylabel("Adjusted Close Price")
plt.xticks(range(0, len(date_lst), 30), date_lst[0::30])

plt.xticks(rotation=45)
plt.legend()

plt.show()
```
